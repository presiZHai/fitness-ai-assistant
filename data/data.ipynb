```python
import pandas as pd
import json
```

```python
# Load the CSV file into a DataFrame
fitness = pd.read_csv('data_clean.csv')
```

```python
# Drop the 'id' column
#fitness = fitness.drop(columns=['id'])
```

```python
# Convert the DataFrame to a dictionary format to resemble the desired JSON structure
fitness_json = fitness.to_dict(orient='records')
```

```python
# Wrap the list in a dictionary with the key "fitness"
fitness_data = {"fitness": fitness_json}
```

```python
# Convert the dictionary to a JSON string
fitness_json_str = json.dumps(fitness_data, indent=2)

print(fitness_json_str)
```

```python
# Save the dictionary to a JSON file
with open('fitness_data.json', 'w') as json_file:
    json.dump(fitness_data, json_file, indent=2)

print("JSON file has been saved successfully.")
```
